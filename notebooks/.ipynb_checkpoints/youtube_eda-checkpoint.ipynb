```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/youtube_videos.csv")
df.head()
```

```python
df["engagement_rate"] = ( (df["likes"] + df["comments"]) / df["views"] ) * 100
df.sort_values("engagement_rate", ascending  =  False).head(5)
```

```python
# views by category
category_wise_views = df.groupby("category")["views"].sum().reset_index(name = "total_views")
category_wise_views.sort_values("total_views", ascending = True).style.hide(axis = 'index')
```

```python
category_wise_views.plot(kind = "bar", x = "category")
plt.title("Category wise total views")
plt.xlabel("Category")
plt.ylabel("Total Views")
plt.show()
```

```python
# Best publishing day
best_publish_days = df.groupby("publish_day")["views"].sum().reset_index(name = "total_views")
best_publish_days.sort_values("total_views", ascending = False).style.hide(axis = "index")
```

```python
best_publish_days.plot(kind = "bar", x = "publish_day")
plt.title("Best days to publish")
plt.xlabel("Days")
plt.ylabel("Total Views")
plt.show()
```

```python
sns.scatterplot(x="views", y="subscribers_gained", hue="category", data=df)
plt.title("Subscribers vs Views by Category")
plt.show()
```

```python
category_wise_subscribers = df.groupby("category")["subscribers_gained"].sum().reset_index(name = "total_subscribers")
category_wise_subscribers.sort_values("total_subscribers", ascending = False).style.hide(axis = 'index')
```
